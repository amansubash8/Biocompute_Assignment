```python
import h5py

# --- CHANGE THIS LINE to your file name ---
filepath = '/Users/aman/Biocompute_assignment/FAQ32498_pass_09083b73_0.fast5' 
# ----------------------------------------

first_read_id = 'read_000a6116-4aac-401e-8fe7-4a24c774128a'
path_to_explore = f'{first_read_id}/Analyses'

def print_hdf5_structure(group, prefix=''):
    """A helper function to print the full structure."""
    for key in group.keys():
        item = group[key]
        path = f'{prefix}/{key}'
        print(path)
        if isinstance(item, h5py.Group):
            # It's a folder, so we look inside it
            print_hdf5_structure(item, prefix=path)

try:
    f = h5py.File(filepath, 'r')
    
    print(f"--- Recursively exploring: {path_to_explore} ---")
    
    analyses_group = f[path_to_explore]
    print_hdf5_structure(analyses_group, prefix=path_to_explore)
        
    f.close()

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import pod5
import numpy as np

# --- Set this to your .pod5 file path ---
filepath = '/Users/aman/Biocompute_assignment/control_rep2.pod5'
# ----------------------------------------

try:
    print(f"Opening file: {filepath}...")
    
    # Open the .pod5 file in read mode ('r')
    with pod5.Reader(filepath) as reader:
        
        print("File opened successfully.")
        
        # Get all read_ids.
        read_ids = list(reader.read_ids)
        print(f"Total number of reads found: {len(read_ids)}")
        
        if len(read_ids) > 0:
            # --- Get the very first read from the file ---
            first_read_id = read_ids[0]
            print(f"\n--- Inspecting the first read ---")
            print(f"Read ID: {first_read_id}")
            
            # --- THIS IS THE CORRECTED LINE ---
            # The correct argument is 'selection'
            first_read = next(reader.reads(selection=[first_read_id]))
            
            # --- 1. Check for Signal ---
            signal = first_read.signal
            if signal is not None:
                # .signal is a numpy array, so we use .shape
                print(f"Signal found. Length: {signal.shape[0]} data points")
            else:
                print("Signal NOT found.")
                
            # --- 2. Check for Analyses ---
            print(f"Run Info: {first_read.run_info.experiment_name}")
            print(f"Pore Info: {first_read.pore.channel}")
            
            print("\n--- CONCLUSION ---")
            print("This .pod5 file contains raw signal and metadata.")
            print("It does NOT contain the 'Analyses' or 'ModifiedBaseEvents' groups.")
            print("This confirms the labels (answers) are in a separate file.")
            
        else:
            print("No reads found in this file.")

except Exception as e:
    print(f"An error occurred: {e}")
```
